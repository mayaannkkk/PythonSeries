```python
import random
jackpot = random.randint(1,100)

guess = int(input("Enter Number:"))
count = 1
while guess != jackpot:
  if guess < jackpot:
    print("guess, higher!")
  else:
    print("guess, Lower!")

  count += 1
  guess = int(input("Enter Number"))

else:
  print("You guess it right!")
  print("Attempts", count )
```

```python
result = 0

num = int(input("Enter Num:"))
fact = 1
for i in range(1,num+1):
  fact = fact * i
  result += i / fact
print(result)
```
