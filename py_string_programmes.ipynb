```python
s = input("Enter Email:")

for i in range(0,len(s)):
  if s[i] == '@':
    break
  else:
    print(s[i],end='')

# simple one

idx = s.index("@")
print(s[0:idx])
```

```python
s = 'dhummachale'

flag = True
for i in range(0,len(s) // 2):
  if s[i] != s[len(s) - i - 1]:
    flag = False
    print("Not a plaindrome")
    break

if flag:
  print("Palindrome")
```
